```python
from dotenv import load_dotenv
from langchain_google_genai import ChatGoogleGenerativeAI

load_dotenv()
```

```python
llm= ChatGoogleGenerativeAI(model="gemma-4-31b-it")

response = llm.invoke("How many moons does jupiter have?")
print(response.text)
```

```python
response = llm.invoke([
    ["system", "You are a helpful assistant that answers in short one line"],
    ["human", "How many moons does jupyter have?"]
])
print(response.text)
```

**Temperature**

```python
prompt = "Write a single tagline for a quirky coffee shop. only output the tagline,nothing else."

llm= ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=0)

response = llm.invoke(prompt)
print(response.text)
```

```python
llm= ChatGoogleGenerativeAI(model="gemini-2.5-flash", temperature=1)

response = llm.invoke(prompt)
print(response.text)
```

**Call LLM using Groq Model**

```python
from langchain_groq import ChatGroq
```

```python
llm = ChatGroq(model="qwen/qwen3-32b", reasoning_format="parsed")

messages = [
    (
        "system",
        "You are a helpful assistant that gives a short answer to a question",
    ),
    (
        "human",
        "How many moons does jupiter have?"
    ),
]
ai_msg = llm.invoke(messages)
print(ai_msg.content)
```
